##Mini Project - Multiple Plots

```python
import pandas as pd
unrate=pd.read_csv("C:\\Users\\91789\\Desktop\\unrate.csv")
unrate
```

```python
unrate["DATE"]=pd.to_datetime(unrate["DATE"])
unrate.head(12)
```

```python
import matplotlib.pyplot as plt
```

```python
x_values=unrate["DATE"].head(12)
y_values=unrate["VALUE"].head(12)
plt.xlabel("Months")
plt.ylabel("Unemployment Rate")
plt.suptitle("Monthly Unemployment Trends,1948")
plt.xticks(rotation=90)
plt.plot(x_values,y_values)
plt.show()
```

```python
fig=plt.figure(figsize=(12,12))
ax1=fig.add_subplot(5,1,1)
ax1.plot(unrate['DATE'][0:12], unrate['VALUE'][0:12],c="red") 
ax1.set_title("Monthly Employment Trends,1948")
ax1.set_xlabel("Month")
ax1.set_ylabel("Unemployment Rate")

ax2=fig.add_subplot(5,1,2)
ax2.plot(unrate['DATE'][12:24], unrate['VALUE'][12:24],c="blue")
ax2.set_title("Monthly Employment Trends,1949")
ax2.set_xlabel("Month")
ax2.set_ylabel("Unemployment Rate")


ax3=fig.add_subplot(5,1,3)
ax3.plot(unrate['DATE'][24:36], unrate['VALUE'][24:36],c="green")
ax3.set_title("Monthly Employment Trends,1950")
ax3.set_xlabel("Month")
ax3.set_ylabel("Unemployment Rate")

ax4=fig.add_subplot(5,1,4)
ax4.plot(unrate['DATE'][36:48], unrate['VALUE'][36:48],c="orange")
ax4.set_title("Monthly Employment Trends,1951")
ax4.set_xlabel("Month")
ax4.set_ylabel("Unemployment Rate")

ax5=fig.add_subplot(5,1,5)
ax5.plot(unrate['DATE'][48:60], unrate['VALUE'][48:60],c="black")
ax5.set_title("Monthly Employment Trends,1952")
ax5.set_xlabel("Month")
ax5.set_ylabel("Unemployment Rate")

fig.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
unrate['MONTH'] = unrate['DATE'].dt.month
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'],c="red")
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'],c="blue")
plt.plot(unrate[24:36]['MONTH'], unrate[24:36]['VALUE'],c="green")
plt.plot(unrate[36:48]['MONTH'], unrate[36:48]['VALUE'],c="orange")
plt.plot(unrate[48:60]['MONTH'], unrate[48:60]['VALUE'],c="black")
plt.xlabel("Month,Integer")
plt.ylabel("Unemployment Rate,Percent")
plt.legend(loc="upper left",labels=["1948","1949","1950","1951","1952"])
plt.title("Monthly Unemployment trends,1948-1952")

plt.show()
```
